# Store Data in CSV format

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://books.toscrape.com/')
data = BeautifulSoup(response.text,'html.parser')
```

```python
b1 = data.find(class_ = 'product_pod')
b1
```

```python
base_url= 'https://books.toscrape.com/'
b1_url = base_url+b1.h3.a['href']  # this is the url of first book . 
print(b1_url)
```

```python
# load the html of first book . 
response = requests.get(b1_url)

data = BeautifulSoup(response.text,'html.parser')
data
```

```python
title = data.h1.string
price = data.find(class_='price_color').string
quantity = data.find(class_= "instock availability")
quantity = quantity.contents[-1].strip()
print(quantity)
print(price)
```

```python
print(title)
print(b1_url)
print(price)
print(quantity)
```

```python
import re

quantity = int(re.search('\d+',quantity).group())  # this line for remove all the other unwanted word like in stock , availablit as  you shown up. 
print(quantity)
```

```python
price = float(re.search('[\d.]+',price).group())
print(price)
```

```python
print(title)
print(b1_url)
print(price)
print(quantity)
```

```python
book_details = []
book_details.append([title,b1_url,price,quantity])
book_details
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(book_details, columns = ['Title', 'URL','Price','Stocks'])
df
```

```python
df.to_csv('bookswebScrapping.csv')
```
